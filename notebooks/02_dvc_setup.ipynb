```python
# Cell 5: Create cleaning script using Python
!mkdir -p src

# Write the file using Python
with open('src/clean_data.py', 'w') as f:
    f.write('''import pandas as pd
import numpy as np

df = pd.read_csv('data/insurance_data.csv')

# Add derived columns
df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
df['Margin'] = df['TotalPremium'] - df['TotalClaims']
df['HasClaim'] = (df['TotalClaims'] > 0).astype(int)

# Save
df.to_csv('data/cleaned_insurance_data.csv', index=False)
print(f"Cleaned: {{df.shape}}")
''')

print(" clean_data.py created")
```

```python
# Cell 6: Create pipeline and run
with open('dvc.yaml', 'w') as f:
    f.write('''
stages:
  clean:
    cmd: python src/clean_data.py
    deps:
      - data/insurance_data.csv
      - src/clean_data.py
    outs:
      - data/cleaned_insurance_data.csv
''')

print(" dvc.yaml created")

!dvc repro

!git add dvc.yaml src/clean_data.py data/cleaned_insurance_data.csv.dvc 2>/dev/null
!git commit -m "add cleaning pipeline"

print(" Pipeline complete!")
```

```python
# Cell 7: Create second version
with open('src/feature_eng.py', 'w') as f:
    f.write('''import pandas as pd
import numpy as np

df = pd.read_csv('data/cleaned_insurance_data.csv')

# Add features if columns exist
if 'RiskScore' in df.columns:
    df['RiskCategory'] = pd.cut(df['RiskScore'], bins=[0, 33, 66, 100], labels=['Low', 'Medium', 'High'])
if 'Age' in df.columns:
    df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 50, 100], labels=['Young', 'Middle', 'Senior'])

df.to_csv('data/insurance_data_v2.csv', index=False)
print(f"Version 2: {{df.shape}}")
''')

print(" feature_eng.py created")

!python src/feature_eng.py
!dvc add data/insurance_data_v2.csv
!git add data/insurance_data_v2.csv.dvc src/feature_eng.py
!git commit -m "create version 2"
!dvc push

print(" Version 2 complete!")
```
